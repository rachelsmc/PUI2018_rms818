# use groupby to count the building per LandUse code 

```python
# counting building per landuse code for manhattan
mngrouped = mn.groupby(by='LandUse').count().reset_index()
mngrouped.head()
```

```python
# group BK by land use
bkgrouped = bk.groupby(by='LandUse').count().reset_index()
bkgrouped.head()
```

# convert the landuse to integer 

```python
# convert landuse type to integer for brooklyn
bkgrouped['LandUse'] = bkgrouped['LandUse'].astype(int)
bkgrouped.head()
```

```python
# convert landuse type to integer for manhattan
mngrouped['LandUse'] = mngrouped['LandUse'].astype(int)
mngrouped.head()
```

# merge the dataframes on land use 

```python
# merge the dataframes on LandUse
both = mngrouped.merge(bkgrouped, how='outer', on='LandUse')
both.columns
```

# rename things so that you have descriptive names

## rename the columns with descriptive names. 


```python
# renaming columns of dataframe
both.rename(index=str, columns={"ZipCode_x": "mn", "ZipCode_y": "bk"}, inplace=True)
both
```

## find online the LandUse data dictionary and replace code w the verbose description of it


```python
# set up Land Use values to descriptive tag
both['LandUse'] = both['LandUse'].map({1: 'One &Two Family Buildings',
                                       2: 'Multi-Family Walk-Up Buildings',
                                       3: 'Multi-Family Elevator Buildings',
                                       4: 'Mixed Residential & Commercial Buildings',
                                       5: 'Commercial & Office Buildings',
                                       6: 'Industrial & Manufacturing',
                                       7: 'Transportation & Utility',
                                       8: 'Public Facilities & Institutions',
                                       9: 'Open Space & Outdoor Recreation',
                                       10: 'Parking Facilities',
                                       11: 'Vacant Land'})
both
```

# Create a feature (a column, a series) with the ratio of number of buildings in BK vs MN and find out in which LandUse class Manhattan has more buildings than BK 

```python
# calculate the ratio of BK to Man buildings per LandUse category
both['LUratio'] = both['mn']/both['bk']
both
```

```python
# defined function that returns True if the ratio is larger than 1, False otherwise
def LandUseRatio(a):
    if a > 1:
        return 'True'
    else:
        return 'False'

# used series.apply() to apply function to LUratio column
both['Manhattan wins'] = both.LUratio.apply(LandUseRatio)
both
```

# Comment on what you found out. Does it make sense? 

Yes it makes sense. Land area of Brooklyn is about three times that of Manhattan, so it is expected that there are more buildings and infrastructure in Brooklyn, so the LUratio is likely to be less than 1. There is an exception for multi-family elevator buildings, which could because many Manhattan buildings are high-rise buildings, hence the need for elevators. Meanwhile in Brooklyn, buildings tend to be low-rise and lower, hence the lack of elevators.
